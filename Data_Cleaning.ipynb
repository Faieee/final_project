```python
#add magic command
%matplotlib inline 

# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import statistics
```

```python
# File to Load 
census_data_to_load = os.path.join('census_data.csv')

# Read the  Data and store into a Pandas DataFrame
census_data_df = pd.read_csv(census_data_to_load)
census_data_df
```

```python
# Determine data types 
census_data_df.dtypes
```

```python
# Determine workclass level

workclass_types = census_data_df.set_index(["workclass"])["income"]
workclass_types
```

```python
# Calculate averages for the desired columns.

State-gov_salary = census_data_df.groupby(["state-gov"]).mean()["Average income"]
self-emp-not-inc_salary = census_data_df.groupby(["self-emp-not-inc"]).mean()["Average income"]
```

```python
# group by workclass category
workclass_count = census_data_df.groupby('workclass')['education'].count()
print(workclass_count)
```

```python
# Determine education level

education_level_types = census_data_df.set_index(["education"])["income"]
education_level_types
```

```python
# group by education category
Education_level_count = census_data_df.groupby('education')['education'].count()
print(Education_level_count)
```
