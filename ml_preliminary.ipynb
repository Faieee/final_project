```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('census_data.csv')
df
```

```python
df.education.unique()
```

```python
df.groupby(df['hours_per_week']).count().head(40)
```

```python
# choose data 

le = LabelEncoder()
df_raw = df[['age', 'workclass', 'education_num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'income']]

df_raw
```

```python
# encode data

df_encoded = df_raw.copy()

for column in df_raw.columns:
    df_encoded[f'{column}'] = le.fit_transform(df_encoded[f'{column}'])
```

```python
df_encoded
```

```python
# prepare ml data

X = df_encoded.drop(columns='income')
y = df_encoded['income']
```

```python
# split data

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```python
classifier = LogisticRegression(solver='lbfgs', max_iter=500)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
